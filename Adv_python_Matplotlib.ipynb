```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]

plt.plot(x,y)
plt.title('Test Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]

plt.plot(x,y)
plt.plot(x,z)
plt.title('Test Graph')
plt.legend(['This is y','This is z'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
